```python
import os
import pandas as pd
```

```python
files_xls = [f for f in files if f[-3:] == 'xls']
files_xls
```

```python
df = pd.DataFrame()
```

```python
for f in files_xls:
    data = pd.read_excel(f, 'Sheet 1')
    df = df.append(data)
```

```python
df = df.iloc[8:]
df
```

```python
df.rename(columns={'Unnamed: 3': 'Total', 'Unnamed: 2': 'Category','Unnamed: 4': 'April15', 'Unnamed: 5': 'May15',
                   'Unnamed: 6': 'June15', 'Unnamed: 7': 'July15','Unnamed: 8': 'August15', 'Unnamed: 9': 'Sept15',
                   'Unnamed: 10': 'Oct15', 'Unnamed: 11': 'Nov15','Unnamed: 12': 'Dec15', 'Unnamed: 13': 'Jan16',
                   'Unnamed: 14': 'Feb16', 'Unnamed: 15': 'March16','Unnamed: 16': 'Totalsum', 'Monthly Progess Report'
                   : 'Id Number'
                    }, inplace=True)
```

```python
df
```

```python
df=df[df['Total'].notnull()]
df
```

```python
df=df.fillna(method='ffill')
df
```

```python
some_values={ "TOTAL","Total"}
df=df.loc[df['Total'].isin(some_values)]
df
```

```python
df1=df.drop(df.columns[[0,3]], axis=1)
```

```python
df1=df1.drop(df1.columns[[14]], axis=1)
```

```python
df2=df1.groupby(['Category','Monthly Progress Report']).sum()
df2
```
